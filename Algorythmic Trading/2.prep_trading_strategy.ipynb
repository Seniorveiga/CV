# Preparing the data for the Trading Strategy

We have obtained this data through the Yahoo! Finance integration. 

Now we are going to add some features to our strategy to make it better.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv", index_col="Date", parse_dates = ["Date"])
data
```

**Adding the feature "Change in Trading Volume (log)"**

We calculate it by taking the volume and dividing it by the volume of the previous day (That´s why the first row does not have any values).

```python
data["vol_ch"] = np.log(data["Volume"].div(data["Volume"].shift(1)))
data
```

## Exploratory Data Analysis: Financial Returns and Trading Volume (1)

**Question 1: Is there any relationship between price changes and volume changes?**

Can we deduce that change in volume derive in price changes?

```python
plt.scatter(x = data["vol_ch"], y = data["returns"])
plt.xlabel("Volume Change")
plt.ylabel("Returns")
plt.show()
```

**Discretize columns returns and vol_ch: Sort and recognize each in 10 equal sized buckets**

```python
pd.qcut(data["returns"], q=10)
```

This will provide us an idea of how the column *"returns"* is distributed in 10 groups.
We divide it by a dummy variable with 10 values from -5 to 5.

```python
data["ret_cat"] = pd.qcut(data["returns"], q=10, labels = [-5,-4,-3,-2,-1,1,2,3,4,5])
```

```python
data
```

Applying the same concept to *Volume_Change*:

```python
data["vol_cat"] = pd.qcut(data["vol_ch"], q=10, labels = [-5,-4,-3,-2,-1,1,2,3,4,5])
```

```python
data
```

We create a crosstab() to have an idea of how much does it affect one another: 
- If you have a lot of values in a certain spot that means that the pattern implies that movement.


```python
matrix = pd.crosstab(data["vol_cat"], data["ret_cat"])
matrix
```

Using a heatmap to visualize it:

```python
import seaborn as sns
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(matrix, cmap = "RdYlBu_r", annot = True, robust = True, fmt = ".0f")
plt.show()
```

### Conclusions

- Extreme cases in trading volume trigger extreme (positively and negatively) returns.
- Low to moderate changes in the trading volume occur frequently with low to moderate returns.

## Exploratory Data Analysis: Financial Returns and Trading Volume (2)

**Can we use return/vol_ch clusters to (partly) forecast future returns?**

```python
data
```

With this information right there, we try to forecast the return of the next bar using the *.shift()* methods. We use this because we are forecasting and not calculating on the present moment the return and the volume.

```python
data["vol_cat"].shift()
```

```python
matrix = pd.crosstab(data["vol_cat"].shift(),data["ret_cat"].shift(),
                    values = data["returns"], aggfunc= np.mean)
matrix
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(matrix, cmap = "RdYlBu", annot = True, robust = True, fmt = ".5f")
plt.show()
```

**The combination of very low volumes with very high returns will produce declines, it is a Contrarian (Reversal Signal)**.

**The combination of very low (negative) returns with very low volumes is a Contrarian (Reversal Signal) -> Prices will rise**.

## Formulating a long-only Price/Volume Trading Strategy

Plan: Make use of the Extremely High (positive) returns and Decreases in Volume Cluster (falling prices)

**Implementation**
- Buy **BTC/USDT** in the spot market and Hold...
- Until there is a signal to sell -> Sell and go neutral
- Buy back when the signal to sell has disappeared

```python
data
```

```python
data["position"] = 1    # Indica la trading position -> long(1) en todas las barras es Buy-and-Hold
data
```

**Sell and go neutral (position = 0) if the returns are very high (cond1) and vol_ch is negative (cond2)**

```python
# getting returns threshold for very high returns (>= 90th percentile)

return_thresh = np.percentile(data["returns"].dropna(),90)
return_thresh
```

```python
cond1 = data["returns"] >= return_thresh
cond1
```

Getting *"vol_ch"* for (moderate) Volume Decreases (between 5th and 20th percentile)


```python
volume_thresh = np.percentile(data["vol_ch"].dropna(),[5,20])
volume_thresh
```

```python
cond2 = data["vol_ch"].between(volume_thresh[0],volume_thresh[1])
cond2
```

```python
data.loc[cond1 & cond2, "position"] = 0
```

```python
data
```

Now we can see the number of times we will be in a **LONG** position compared to a **NEUTRAL** position.

```python
data["position"].value_counts()
```

```python
data.loc[:,"position"].plot(figsize = (12,8))
plt.show()
```

## Strategy Backtest

```python
data
```

To know how much we are earning, just multiply the returns column with the position column.

**Important**: The positions refer to previous values by the *shift()*. Let's create a new column to tell us the position for the next step:

```python
data["strategy"] = data["position"].shift(1) * data["returns"]
```

```python
data
```

In case of neutral positions:

```python
data.loc["2017-08-19"]
```

We calculate the multiple of  the BuyAndHold Strategy

```python
data[["returns","strategy"]].sum().apply(np.exp)
```

This means that with the BuyAndHold the strategy would have generated $12.5 while with the other one it would have generated $35.

```python
# Normalized price with base 1 strategy
data["cstrategy"] = data["strategy"].cumsum().apply(np.exp)
```

```python
data
```

```python
data[["creturns","cstrategy"]].plot(figsize=(12,8),fontsize=12)
plt.show()
```

```python
tp_year = 24 * 365.25    # 1h trading periods per year
tp_year
```

If we obtain a higher mean than the BuyAndHold and if we obtain a lower standarad deviation than the BuyAndHold, we will have a more profitable strategy with less risk associated:

```python
ann_mean = data[["returns","strategy"]].mean() * tp_year
ann_mean
```

```python
ann_std = data[["returns","strategy"]].std() * np.sqrt(tp_year)
ann_std
```

And, in fact, the Sharpe Ratio is higher:

```python
sharpe = (np.exp(ann_mean) - 1)/ ann_std
sharpe
```

It is above 1 which is a very good signal.

## Trading Costs

Nevertheless, we should always consider the trading costs as they will provide the real value in our strategy, so it would unrealistic to not specify their value.

```python
data
```

```python
data["position"].value_counts()
```

Most of the time we are in a Long position.

```python
data["position"].diff().fillna(0).abs()    # Absolute Change in Position
```

```python
data["trades"] = data["position"].diff().fillna(0).abs()
```

```python
data["trades"].value_counts()
```

We have done:

-> 718 operations (From long to neutral and viceversa)

-> Ever trade has costs that should not be ignored!!!

-> Trading costs **must be included** in backtesting!!!

```python
commissions = 0.00075    #Binance comission 0.075%
commissions
```

```python
other = 0.0001  # Proportional cost for bid-ask spread 
other
```

```python
# CTotal trading costs
ptc = np.log(1 - commissions) + np.log(1 - other)
ptc
```

```python
data
```

```python
data["strategy_net"] = data["strategy"] + data["trades"] * ptc #strategy returns net of costs
```

```python
data["cstrategy_net"] = data["strategy_net"].cumsum().apply(np.exp)
```

```python
data
```

```python
data[["creturns","cstrategy","cstrategy_net"]].plot(figsize=(12,8))
plt.show()
```
